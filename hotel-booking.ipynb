### 1. ลูกค้าใหม่มีการร้องขอพิเศษจากโรงแรมมากกว่าลูกค้าเก่า (ตั้งแต่ทำการจองจนถึง Check-Out) เป็นเรื่องบังเอิญหรือไม่

**เป้าหมาย**: อยากทราบว่าตั้งแต่จองจนถึง Check-Out ลูกค้าใหม่มีการร้องขอพิเศษกับทางโรงแรมมากกว่าลูกค้าเก่านั้นเป็นเพียงความบังเอิญ หรือเป็นเพราะความแตกต่างระหว่างคนที่เคยจองกับทางโรงแรมนี้แล้วกับคนที่ไม่เคยจองกันแน่

- ลูกค้าเก่า หมายถึง รายชื่อผู้ทำการจองมีรายชื่อในระบบของโรงแรมแล้ว (is_repeated_guest = 1)
- ลูกค้าใหม่ หมายถึง รายชื่อผู้ทำการจองไม่มีรายชื่อในระบบของโรงแรมมาก่อน (is_repeated_guest = 0)

คัดกรองเอาเพียงคอลัมน์ที่จำเป็น และคัดลอกตารางเพื่อเก็บไว้ดำเนินการต่าง ๆ โดยไม่ให้กระทบกับตารางข้อมูลหลัก และเนื่องจากต้องการพิจารณาตั้งแต่จองถึง Check-Out จึงคัดเลือกเอาเพียงข้อมูลที่สถานะล่าสุดคือการ Check-Out เท่านั้น

```python
first = df.loc[df.reservation_status == 'Check-Out', ['is_repeated_guest', 'total_of_special_requests']]

# เปลี่ยนค่า is_repeated_guest จาก 0 และ 1 เป็น New Guest และ Repeated Guest ตามลำดับเพื่อให้ง่ายต่อการอ่านข้อมูล
repeated_guest = {0 : 'New Guest', 1: 'Repeated Guest'}
first.is_repeated_guest = first.is_repeated_guest.map(repeated_guest)

first
```

Visualize และ สร้างตารางใหม่เพื่อดูภาพรวมของข้อมูล is_repeated_guest และ total_of_special_requests

```python
# กำหนดจำนวนข้อมูลของแต่ละกลุ่ม
num_grp_A1 = num_by_repeated_guest1['Repeated Guest']
num_grp_B1 = num_by_repeated_guest1['New Guest']
num_grp_A1, num_grp_B1
```

```python
# จำนวนการวนซ้ำ
num_loop1 = 2000
```

```python
# list เก็บผลต่างของค่าเฉลี่ยการร้องขอพิเศษ
ls_request_diff1 = []

for i in range(num_loop1):
    # การสุ่มข้อมูลมาสร้าง กลุ่ม A โดยกำหนด random_state เพื่อให้ทุกครั้งที่มีการรันผลลัพธ์ออกมาเหมือนกัน
    grp_A = first.sample(num_grp_A1, random_state= (i*i +7 * num_loop1))

    # สร้างกลุ่ม B โดยนำข้อมูลที่เหลือมาสร้าง
    grp_B = first.loc[ ~first.index.isin( grp_A.index ), :]

    # คำนวณผลต่างของค่าเฉลี่ยของทั้ง 2 กลุ่ม
    mean_request_diff = grp_B.total_of_special_requests.mean() - grp_A.total_of_special_requests.mean()

    # เพิ่มผลต่างของค่าเฉลี่ยของทั้ง 2 กลุ่มไว้ใน ls_request_diff
    ls_request_diff1.append(mean_request_diff)
```

4. คำนวณ P-Value

```python
# นำผลต่างจาก ls_request_diff มาสร้างเป็น Series
request_diff1 = pd.Series(ls_request_diff1)
request_diff1
```

```python
# คำนวณ p-value
p_value_first = (request_diff1 >= real_diff1).sum() / num_loop1
p_value_first
```

5. สรุปผลจากการเปรียบเทียบ p-value

- ถ้าหากคำนวณ p_value แล้วได้ค่าที่ต่ำกว่าระดับนัยสำคัญ หมายความว่าเหตุการณ์ที่เกิดขึ้น เกิดขึ้นค่อนข้างน้อยดังนั้นจึงไม่น่าใช่ความบังเอิญ
    * ปฏิเสธ H<sub>0</sub> ยอมรับ H<sub>1</sub>
- ถ้าหากคำนวณ p_value แล้วได้ค่าที่มากกว่าหรือเท่ากับระดับนัยสำคัญ หมายความว่าเหตุการณ์ที่เกิดขึ้น เกิดขึ้นค่อนข้างบ่อยไม่ได้แปลกแต่อย่างใด ดังนั้นจึงน่าจะเป็นความบังเอิญ
    * ยอมรับ H<sub>0</sub> ปฏิเสธ H<sub>1</sub>

```python
if (p_value_first < alpha1):
    print(problem1 , 'ไม่ใช่เรื่องบังเอิญ แต่เป็นเพราะความแตกต่างระหว่างลูกค้าเก่าและลูกค้าใหม่')
else :
    print(problem1 , 'เป็นเพียงเรื่องบังเอิญ')
```

#### 3.2 ทำการสร้างและฝึกสอนโมเดล

เริ่มทำการสร้างและฝึกสอนโมเดล โดยใช้ train_test_split ในการแบ่งข้อมูลเป็นชุดฝึกและชุดทดสอบ โดยมีพารามิเตอร์ดังนี้

- X : ข้อมูลที่ใช้ในการฝึกและทดสอบโมเดล (feature variables)
- y : ตัวแปรตามที่ต้องการทำนาย (target variable)
- test_size : ขนาดของชุดข้อมูลทดสอบ คิดเป็นร้อยละ กำหนดให้เป็น 0.2
- random_state : ค่าที่ใช้ในการกำหนดการสุ่มข้อมูล กำหนดให้เป็น 920

```python
# แยก target variable (y) ออกจาก training set (X)
y = dummy_ml.is_canceled
X = dummy_ml.drop('is_canceled', axis = 1)

X.head(5)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=920)
```

```python
# ตรวจสอบขนาดของชุดข้อมูล
X_train.shape[0], X_test.shape[0], y_train.shape[0], y_test.shape[0]
```

ใช้ Grid Search Cross-Validation เพื่อหาพารามิเตอร์ที่ดีที่สุดสำหรับโมเดล โดยใช้พารามิเตอร์ที่กำหนดใน param_grid ได้แก่
- max_depth คือ ความลึกของต้นไม้ตัดสินใจ
- min_samples_split คือ จำนวนตัวอย่างขั้นต่ำที่ต้องมีในโหนดใดโหนดหนึ่งก่อนที่จะแยกโหนดนั้นออกเป็นโหนดย่อย

สร้างโมเดลโดยใช้พารามิเตอร์ที่ระบุไว้ และกำหนดจำนวน Fold ของ Cross-Validation เป็น 8 แล้วทำการ Fit ชุดข้อมูลฝึกในโมเดล เพื่อหาค่าพารามิเตอร์ที่ดีที่สุด และเก็บไว้ในตัวแปร best_dt



```python
param_grid = {
    'max_depth': [20, 21, 22, 23],
    'min_samples_split':[2, 3, 5, 6],
}

# Perform Grid Search with Cross-Validation
dt = DecisionTreeClassifier(random_state=132)
grid_dt = GridSearchCV(estimator=dt, param_grid=param_grid, cv=8)

grid_dt.fit(X_train, y_train)

# พิมพ์ best parameters และ best score
print('Best Parameters:', grid_dt.best_params_)
print('Best Score:', grid_dt.best_score_)

# เก็บโมเดล Decision Tree ที่ดีที่สุดที่ได้จากการค้นหาไว้ในตัวแปร best_dt
best_dt = grid_dt.best_estimator_
```

พบว่าพารามิเตอร์ที่ดีที่สุดคือ max_depth เท่ากับ 22 และ min_samples_split เท่ากับ 2 โดยมีค่าความแม่นยำที่ดีที่สุด (best score) คือ 0.8657 หรือประมาณ 86.57%

#### 3.3 ทำการทดสอบประสิทธิภาพของโมเดล

ใช้ชุดข้อมูลทดสอบเพื่อทดสอบประสิทธิภาพของโมเดลโดยการ และแสดงผลลัพธ์ที่สามารถวัดประสิทธิภาพของโมเดลได้ ได้แก่ accuracy score, confusion matrix, กราฟ, classification report และแผนภาพต้นไม้

```python
# Predict โดยใช้ test set
y_pred = best_dt.predict(X_test)

# Model accuracy
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
